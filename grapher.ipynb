```python
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
df = pd.read_csv('history.csv')

df.head()
```

```python
fig, axs = plt.subplots(2,1, figsize=(12,6), sharex=True, layout='tight')
plt.subplot(2,1,1)
plt.plot(df['volume'])
plt.ylim(0, 100)
plt.axhline(y=80, color='r', linestyle='--')
plt.ylabel('Volume [%]', fontsize=16)

plt.subplot(2,1,2)
plt.plot(df['flow_in'])
plt.ylim(0, 15)
plt.ylabel('Vazo de entrada [L/s]', fontsize=16)
plt.xlabel('Tempo (s)', fontsize=16)

#plt.subplot(3,1,3)
#plt.plot(df['flow_out'])
#plt.ylim(0, 15)

```
